```python
import json
import pickle

with open("/home/makilins/Downloads/movies.p","rb") as tmdb_file:
    data = pickle.load(tmdb_file)

    batch_size = 12
    num_batches = len(data)/batch_size
    if len(data)%batch_size != 0:
        #then there's an additional partial batch to account for
        num_batches += 1    

    batches = []
    for batch_num in range(int(num_batches)):
        #print(type(data))
        movs = data[batch_num*batch_size:(batch_num+1)*batch_size]
        #print(movs)
        json_docs = []
        for mov in movs:
            mov_short = {}
            mov_short["id"] = mov["id"]
            mov_short["title"] = mov["title"]
            mov_short["overview"] = mov["overview"]
            mov_short["tagline"] = mov["tagline"]
            mov_short["release_date"] = mov["release_date"]
            mov_short["budget"] = mov["budget"]
            mov_short["revenue"] = mov["revenue"]
            mov_short["vote_count"] = mov["vote_count"]
            mov_short["vote_average"] = mov["vote_average"]
            mov_short["runtime"] = mov["runtime"]
            mov_short["production_companies"] = mov["production_companies"]
            mov_short["genres"] = [genre["name"] for genre in mov["genres"]]
            mov_short["production_companies"] = [prod_co["name"] for prod_co in mov["production_companies"]] 

            # Нужно эластику для индексирования
            addCmd = {"index": {"_index": "tmdb", "_id": mov_short["id"]}} 
            json_docs.append(json.dumps(addCmd))
            json_docs.append(json.dumps(mov_short))
        batches.append("\n".join(json_docs)+"\n")

    
```

```python
import pprint 
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(batches[0])
```

```python
import requests

elasticSearchUrl = "http://localhost"
```

```python
for i, batch in enumerate(batches):
    r = requests.post(elasticSearchUrl + "/_bulk", data=batch, headers={"Content-Type": "application/x-ndjson"})
    print(i, r.text)
```

```python
# Фильтр - берем те фильмы, чей бюджет меньше 100000000

search = json.dumps(
    {
    "query": { 
        "bool": {
            "filter": [{
                "range": {
                    "budget": {
                        "gte": "100000000"
                        }
                    }
                }]
            }
        }
    }
)

print(requests.get("http://localhost:9200/tmdb/_doc/_search?pretty", 
                   data=search,
                   headers={"Content-Type": "application/json"}).json())
```

```python
# Полнотекстовый поиск - ищем фильмы Marvel

search = json.dumps(
    {
    "query": {
        "match" : {
            "production_companies" : {
                "query" : "Marvel"
            }
        }
    }
}
)

print(requests.get("http://localhost:9200/tmdb/_doc/_search?pretty", 
                   data=search,
                   headers={"Content-Type": "application/json"}).json())
```
